```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris_ = load_iris()
iris= pd.DataFrame(data=iris_.data, columns=iris_.feature_names)
iris['species'] = iris_.target_names[iris_.target]
```

```python
#Histograms
iris.hist(figsize=(12,8), bins=20, color='#02bef7', edgecolor='black')
plt.suptitle('Histograms of Iris Dataset Columns', x=0.5, y=1.02, ha='center', fontsize='large')
plt.tight_layout()
plt.show()
```

```python
#Scattering Plots
plt.figure(figsize=(12,8))
plt.subplot(2, 2, 1)
plt.title('Scatter Plot: Sepal Length vs. Sepal Width')
plt.scatter(iris['sepal length (cm)'], iris['sepal width (cm)'], color='#07f767', edgecolor='black')


plt.subplot(2, 2, 2)
plt.title('Scatter Plot: Sepal Length vs. Petal Length')
plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'], color='#f55ff0', edgecolor='black')


plt.subplot(2, 2, 3)
plt.title('Scatter Plot: Petal Length vs. Petal Width')
plt.scatter(iris['petal length (cm)'], iris['petal width (cm)'], color='#02bef7', edgecolor='black')


plt.subplot(2, 2, 4)
plt.title('Scatter Plot: Sepal Width vs. Petal Width')
plt.scatter(iris['sepal width (cm)'], iris['petal width (cm)'], color='yellow', edgecolor='black')


plt.suptitle('Scatter Plots of Iris Dataset Columns', x=0.5, y=1, fontsize='large')
plt.tight_layout()
plt.show()
```

```python
# Mean , Median , Mode , Standard Deviation Values for each
for i in iris.columns :
  if i!='species':
    print(i)
    print('mean:',np.mean(iris[i]))
    print('median:',np.median(iris[i]))
    print('mode:',iris[i].mode().values[0])
    print('Standard deviation:',np.std(iris[i]))
    print('\n')
```
