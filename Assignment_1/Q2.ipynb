```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
#Question1
arr = np.random.randint(1,100,size=(4,4,4))
print('array: \n',arr)
mean = np.mean(arr)
print('mean: ',mean)
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]
print('IRIS dataframe: \n',iris_df,'\n')
iris_df.to_csv('iris.csv',index=False)
```

```python
#Question2
#reading csv file
iris_file = pd.read_csv('iris.csv')
#filtering rows with sepal length (cm) > 5.0
print(iris_file.loc[(iris_file['sepal length (cm)']>6.0)])
```

```python
#Question3
a=[1,2,3,4]
b=[3,6,9,12]

plt.title('Q3-Graph',fontdict={'family':'serif','fontsize':15})
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(a,b,'g.--',label='y=3x',markersize=15,markeredgecolor='r')
plt.legend(loc='upper left')
plt.show()
```

```python
#Question4
#adding a nan row
df1 = iris_df.copy()
df1.loc[len(iris_df)] = np.nan
print('df1: \n',df1)
#replacing nan values with mean of the columns
for i in df1.columns:
  df1.fillna(df1.mean(numeric_only=True), inplace=True)
print('df1: \n',df1)
```

```python
#Question5
matrix1 = np.random.randint(1,10,size=(2,3))
matrix2 = np.random.randint(1,10,size=(3,2))
print('matrix1: \n',matrix1)
print('matrix2: \n',matrix2)
dot_prod_matrix = np.dot(matrix1,matrix2)
print('Resultant matrix: \n',dot_prod_matrix)
det = np.linalg.det(dot_prod_matrix)
print('determinant: ',det)
```

```python
#Question6
df2 = iris_df.copy()
print('Groupby sum values: \n')
for i in df2.columns :
  if i!='species':
    print(df2.groupby(['species'])[i].sum())

```
